# 3. Outlier removing

```python
plt.figure(figsize=(30,30))
for i, col in enumerate(data.columns.to_list()):
    plt.subplot(5, 5, i + 1)
    plt.hist(data[col], label=col,color='blue')
    plt.title(col)
    plt.tight_layout()
```

Outlier는 없는것 같기에 제거하지 않겠습니다.

# 4. Correlation removing

```python
plt.figure(figsize=(30,30))
sns.heatmap(data.corr(), annot=True, cmap='Blues')
```

연관성이 높아보이는 것은 크게 없기에 제거하지 않겠습니다.
